```python
import pandas as pd
import numpy as np 
import csv 
import matplotlib.pyplot as plt

f = open('data/gender.csv',encoding='cp949')
data = csv.reader(f)
size=[]
name=input('지역 입력:')
for row in data:
    if name in row[0]:
        m=0
        f=0
        for i in range(101):
            
            m+=int(row[i+3])
            f+=int(row[i+106])
        
        break

plt.rc('font', family='AppleGothic')
color =['crimson','darkcyan']
plt.axis('equal')
plt.pie(size,autopct='%.1f%%',colors=color) #autopct='%.1f%% 퍼센트 표시
plt.title(name+'지역의 남녀 성별 비율')
plt.show()
```

```python
#성별 연령별 인구 데이터 꺽은 선 그래프로 실행하기
f = open('data/gender.csv',encoding='cp949')
data = csv.reader(f)

m = []
f = []
dong = input('찾고 싶은 지역의 이름을 알려주세요:')
for row in data:
    if dong in row[0]:
        for i in range(3,104):
            m.append(int(row[i]))
            f.append(int(row[i+103]))
        break

plt.rc('font', family='AppleGothic')
plt.plot(m, color ='r', label='Aug')
plt.plot(f, color ='b', label='Jan')
```

```python
#연령별 남성 인구 - 여성인구 값 막대 그래프로 표현하기 (실행결과)
f = open('data/gender.csv',encoding='cp949')
data = csv.reader(f)

result = []

dong = input('찾고 싶은 지역의 이름을 알려주세요:')
for row in data:
    if dong in row[0]:
        for i in range(3,104):
            result.append(int(row[i])- int(row[i+103]))
        break

plt.rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus']=False
plt.bar(range(101),result,color=['green'])
plt.show()
```
